```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('./data/train.csv')
df.head()
```

```python
df = df.drop(['PassengerId', 'Name', 'Ticket'], axis=1)
sns.heatmap(df.corr(), annot=True)
```

```python
df.head()
```

```python
df['Cabin'] = (df['Cabin'].notnull()).astype('int')
df['Cabin'] = df['Cabin'].fillna(0)
```

```python
df
```

```python
df.describe()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df.describe()
```

```python
df = pd.get_dummies(df)
df.head()
```

```python
from sklearn.model_selection import train_test_split
random_state = 3
df_train, df_test = train_test_split(df, test_size=0.20, random_state=random_state)
```

```python
X_train, y_train = df_train.loc[:, df_train.columns != 'Survived'], df_train['Survived']
X_train.head(), y_train.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
```

```python
from sklearn.linear_model import LogisticRegression

rfc = LogisticRegression(random_state=random_state)
rfc.fit(X_train, y_train)
```

```python
X_test, y_test = df_test.loc[:, df_test.columns != 'Survived'], df_test['Survived']
X_test = scaler.transform(X_test)

rfc.score(X_train, y_train), rfc.score(X_test, y_test)
```
